# Extract data from stockes

```python
df_stocke_utc = df_stocke.tz_convert('UTC') # we match the Timezone with google trends

dates_utc = pd.date_range(start=begin,end=last,freq='d')
dates_utc = dates_utc.tz_localize('UTC') # we create a new datetime with the UTC timezone
len(dates)
len(dates_utc)
h = [x for x in range(24)]

open = np.array([])
close = np.array([])
volume = np.array([])
high = np.array([])
low = np.array([])

for i in dates_utc.date:
  for j in h:
    temp = df_stocke.loc[(df_stocke_utc.index.date == i)&(df_stocke_utc.index.hour == j)]
    temp_open = temp['Open'].to_numpy()
    temp_close = temp['Close'].to_numpy()
    temp_volume = temp['Volume'].to_numpy()
    temp_high = temp['High'].to_numpy()
    temp_low = temp['Low'].to_numpy()

    # Open value #
    if (temp_open.size == 0):
      # no match, so we put an empty value
      open = np.append(open,-1)
    elif (temp_open.size == 1):
      open = np.append(open,temp_open)
    else:
      # more than one element, we don't trust the data so we put an empty value []
      open = np.append(open,-1)

    # Close value #
    if (temp_close.size == 0):
      # no match, so we put an empty value
      close = np.append(close,-1)
    elif (temp_close.size == 1):
      close = np.append(close,temp_close)
    else:
      # more than one element, we don't trust the data so we put an empty value []
      close = np.append(close,-1)

    # Volume value #
    if (temp_volume.size == 0):
      # no match, so we put an null value
      volume = np.append(volume,0)
    elif (temp_volume.size == 1):
      volume = np.append(volume,temp_volume)
    else:
      # more than one element, we don't trust the data so we put an null value []
      volume = np.append(volume,0)

    # High value #
    if (temp_high.size == 0):
      # no match, so we put an empty value
      high = np.append(high,-1)
    elif (temp_high.size == 1):
      high = np.append(high,temp_high)
    else:
      # more than one element, we don't trust the data so we put an empty value []
      high = np.append(high,-1)

    # Low value #
    if (temp_low.size == 0):
      # no match, so we put an empty value
      low = np.append(low,-1)
    elif (temp_low.size == 1):
      low = np.append(low,temp_low)
    else:
      # more than one element, we don't trust the data so we put an empty value []
      low = np.append(low,-1)

    if (i == dates_utc.date[-1]):
      break
print(len(open))
print(len(close))
print(len(high))
print(len(low))
print(len(volume))
print(len(tesla))
```

# Create vector of input to the ANN

```python
# Specify the number of hours into the past from the opening hour that will be used for the ANN
utc_opening_hour = 13
time_past = 24*3
sessionDur=7
nDays = time_past//(utc_opening_hour+1) # computes the number of days in which we will have to start 
                                        # the extraction to avoid dimension problems with the dataframe
nVectors = len(dates_utc)-nDays-1
#computation dimensions of the output array
nCol = len(df_hourly_data.columns)*time_past

inputs = np.array([])
outputs = np.array([])

for i in range(0,nVectors):
    lastRow = 24*(nDays+i)+utc_opening_hour
    iniRow = lastRow-time_past
    inputs = np.append(inputs,df_hourly_data.iloc[iniRow:lastRow,:].to_numpy().flatten())
    outputs = np.append(outputs,np.array(df_hourly_data.iloc[lastRow+sessionDur,0]))

inputs = np.reshape(inputs,[nVectors,5*time_past])
print(inputs)
print(outputs)
print(inputs.shape)
print(outputs.shape)
```

```python
colName = ['feature_{0}'.format(s) for s in range(nCol)]

dataContent = dict()

for i in range(nCol):
  dataContent[colName[i]] = inputs[:,i].tolist()

dataContent['target'] = outputs.tolist()
#dataContent['feature_47']
df_ann = pd.DataFrame.from_dict(dataContent)

df_ann.head()
```

```python
df_forecasting_label = df_forecasting_na.copy()
```

```python
# Parameter: percentage of broker
percentage = 0.0075

temp_clos = df_forecasting_label['close'].to_numpy()
temp_label = np.array([]) # 0 -> fix, 1 -> increase, 2-> decrease

for i in range(temp_clos.size):
  if i == (temp_clos.size-1):
    temp_label = np.append(temp_label,0)
  else:
    incr = (temp_clos[i+1]-temp_clos[i])/(temp_clos[i])
    if incr > percentage: # value increase more than the margin
      temp_label = np.append(temp_label,1)
    elif incr < -percentage:# value decrease less than the margin
      temp_label = np.append(temp_label,2) 
    else:
      temp_label = np.append(temp_label,0)

df_forecasting_label['label'] = temp_label
```
